```python
import pandas as pd
import re
from yahoofinancials import YahooFinancials
```

```python
ticker = "PETR4.SA"
yf = YahooFinancials(ticker)
```

```python
income_statment_data_a = yf.get_financial_stmts("annual", "income")
balance_sheet_data_a = yf.get_financial_stmts("annual", "balance")
cash_flow_data_a = yf.get_financial_stmts("annual", "cash")
```

```python
def yf_to_pd(x):
    x = next(iter(next(iter(x.values())).values()))
    
    df_list = []
    
    for i in x:
        df_list.append(pd.DataFrame.from_dict(i, orient="index"))
    
    return pd.concat(df_list).transpose()
```

```python
petr_is = yf_to_pd(income_statment_data_a)
petr_is.sort_index(axis=1, inplace=True)
```

```python
petr_bs = yf_to_pd(balance_sheet_data_a)
petr_bs.sort_index(axis=1, inplace=True)
```

```python
petr_cf = yf_to_pd(cash_flow_data_a)
petr_cf.sort_index(axis=1, inplace=True)
```

```python
def change_index(n):
    return " ".join(re.findall("[A-Z][^A-Z]*", (n[0].upper() + n[1:])))
```

```python
petr_is.index = petr_is.index.map(change_index)
petr_bs.index = petr_bs.index.map(change_index)
petr_cf.index = petr_cf.index.map(change_index)
```
